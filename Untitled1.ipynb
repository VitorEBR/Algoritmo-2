```python
def divisao_passo_a_passo(dividendo, divisor, casas=10):
    if divisor == 0:
        print("Erro: divisão por zero.")
        return

    print(f"Dividindo {dividendo} por {divisor}\n")

    resto = dividendo
    quociente = ""
    passos = []

    # Parte inteira
    inteiro = resto // divisor
    resto = resto % divisor
    quociente += str(inteiro)

    passos.append(f"{dividendo} ÷ {divisor} = {inteiro} com resto {resto}")

    # Parte decimal
    if resto != 0:
        quociente += "."

    for i in range(casas):
        resto *= 10
        digito = resto // divisor
        novo_resto = resto % divisor

        passos.append(
            f"Passo {i+1}: {resto} ÷ {divisor} = {digito} com resto {novo_resto}"
        )

        quociente += str(digito)
        resto = novo_resto

        if resto == 0:
            break

    # Exibição formatada
    print("=== Passo a passo ===")
    for p in passos:
        print(p)

    print("\nResultado final:")
    print(quociente)
```

```python
divisao_passo_a_passo(22, 7)
```
