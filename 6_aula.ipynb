```python
lista = [50,1,0,60,13,24,99,102,64,78,95,10000,85274,46519465498,488494794651465,44644,31122,55733,555476993663,1444987,112204,333,412]
```

```python
def InsertionSort(lista):
  n=len(lista)
  for i in range(1,n):
    chave=lista[i]
    j=i-1
    while j>=0 and lista[j]>chave:
      lista[j+1]=lista[j]
      j-=1
    lista[j+1]=chave
  return lista
```

```python
Lista_Ordenada = InsertionSort(lista)
```
