```python
def quicksort(lista_aleatoria):
  if len(lista_aleatoria)<=1:
    return lista_aleatoria
  pivo=lista_aleatoria[0]
  menores=[x for x in lista_aleatoria[1:] if x <pivo]
  maiores=[x for x in lista_aleatoria[1:]if x >=pivo]
  return quicksort(menores)+[pivo]+quicksort(maiores)
```
